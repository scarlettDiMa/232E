```r
# Q8a
gene_i = list()
for(i in 1:length(sizes(communities))) { 
    movie_i = (1:vcount(g))[communities$membership == i]
    movie_i_gene = V(g)[movie_i]$Genre
    movie_i_gene = table(movie_i_gene)
    gene_i[[i]] = names(movie_i_gene[which.max(movie_i_gene)])
}
for (i in range) {
    print(sprintf("The the most frequent dominant genre across community %s is %s", i, gene_i[[i]]))
}
```

```r
# Q8b
com_size = 0
total_size = sum(sizes(communities))
all_genre = table(V(g)$Genre)
genre_i = list()
score_i = list()
for(i in 1:length(sizes(communities))) { 
    max_score = 0
    com_size = sizes(communities)[i]
    movie_i = (1:vcount(g))[communities$membership == i]
    movie_i_gene = V(g)[movie_i]$Genre
    movie_i_gene = table(movie_i_gene)
    for (j in 1:length(movie_i_gene)) {
        genre = names(movie_i_gene)[j]
        c = movie_i_gene[[j]]
        p = c / com_size
        q = all_genre[[genre]] / total_size
        score = log(c) * p / q
        if (score > max_score) {
            max_score = score
            score_i[[i]] = max_score
            genre_i[[i]] = genre
        }
    }
}
for (i in range) {
    print(sprintf("The the most frequent dominant genre across community %s is %s with score %s", i, genre_i[[i]], score_i[[i]]))
}
```

```r
# Q8 (b)
com_size = 0
total_size = sum(sizes(communities))
all_genre = table(V(g)$Genre)
genres = list()
scores = list()
for(i in 1:length(sizes(communities))) { 
    max_score = 0
    com_size = sizes(communities)[i]
    movies = (1:vcount(g))[communities$membership == i]
    movies_gene = V(g)[movies]$Genre
    movies_gene = table(movies_gene)
    for (j in 1:length(movies_gene)) {
        genre = names(movies_gene)[j]
        c = movies_gene[[j]]
        p = c / com_size
        q = all_genre[[genre]] / total_size
        score = log(c) * p / q
        if (score > max_score) {
            max_score = score
            scores[[i]] = max_score
            genres[[i]] = genre
        }
    }
}
for (i in range) {
    print(sprintf("The the most frequent dominant genre across community %s is %s with score %s", i, genres[[i]], scores[[i]]))
}
```

```r
find_neighbors = function(g, netg, movie) {
    id = V(g)[which(V(g)$name == movie)]
    nei_ids = get.edgeIDs(netg,id,neighborhood="combined")
    rating_list = c()
    for (nei_id in nei_ids) {
        rate = V(g)[nei_id]$rating
        if (rate != "NA") {
            rating_list = rbind(rating_list, rate)
        }
    }
    occur = as.vector(table(rating_list))
    occur = occur/sum(occur)
    p = occur/sum(occur)
    print(sum(occur))
    x = as.numeric(names(table(rating_list)))
    print(sprintf("most frequent rating of neighbors: %s", x[which.max(p)]))
    print(sprintf("average rating of neighbors: %s", mean(x)))
    plot(x, p, type = "o", main = sprintf("Neighbor Rating Degree Distribution (Movie Id = %s)", movie), xlab = "Degree", ylab = "Frequency", col = "blue")
}
```
